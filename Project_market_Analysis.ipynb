<h3> Replacing Income2 with Levels as (1,2,3,4,5)</h3>

```python
def condition(x):
    if x=="<30k":
        return "1"
    elif x=="30-60k":
        return "2"
    elif x=="60-90k":
        return "3"
    elif x=="90-120k":
        return "4"
    elif x==">120k":
        return "5"
    else:
        return 'NA'
```

```python
df['Income2'] = df['Income2'].apply(condition)
```
